### Conceptos previos:
### ¿Qué es el EDA y por qué es no negociable?
El EDA (Exploratory Data Analysis) es el proceso de comprender la estructura, distribución, relaciones y anomalías de un dataset antes de modelar. Fue formalizado por John Tukey (1977).


Principio fundamental: Un modelo entrenado sobre datos mal comprendidos hereda todos sus problemas. El EDA es la única defensa.

### Conceptos Clave:
### 1. Distribución univariada
La distribución de una variable describe la frecuencia con que toma cada valor. Para ventas esperás asimetría positiva (skewness > 0): muchos días con ventas bajas, pocos con ventas muy altas.

### 2. Outliers
Observaciones que se alejan significativamente del patrón general. En series de tiempo, los outliers pueden ser:

* Aditivos (AO): spike puntual (ej: promoción de un día)
* De nivel (LS): cambio permanente en la media (ej: apertura de nueva tienda)
* Transitorios (TC): efecto que decae gradualmente

### 3. Estacionariedad (concepto central en series de tiempo)
Una serie es estacionaria si su media, varianza y autocovarianza no cambian en el tiempo. La mayoría de las series de ventas reales no son estacionarias → tienen tendencia y estacionalidad.
